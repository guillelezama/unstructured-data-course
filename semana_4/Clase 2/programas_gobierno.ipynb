```python
import requests
import pdfplumber
from io import BytesIO
```

```python
# URLs originales
urls = [
    'https://www.frenteamplio.uy/wp-content/uploads/Bases-Programaticas-2025-2030.pdf',
    'https://alvarodelgado.uy/pdf/programa-gobierno.pdf',
    'https://programapartidocolorado.uy/wp-content/uploads/2024/08/LIBRO-PROGRAMA-UNICO-DE-GOBIERNO-2025-PARTIDO-COLORADO.pdf',
    'https://partidoindependiente.uy/wp-content/uploads/2024/05/Programa.-pdf-1.pdf',
    'https://cabildoabierto.uy/wp-content/uploads/2024/08/Programa-CA-2025-2030.pdf'
]

# Nombres descriptivos
nombres_partidos = [
    'Frente Amplio',
    'Partido Nacional',
    'Partido Colorado',
    'Partido Independiente',
    'Cabildo Abierto'
]
```

```python
# Hacer un loop, para cada elemento de la lista de links, bajarlo, transformarlo a texto, y guardarlo en un diccionario texto_programas: {Frente Amplio: TEXTO DEL PROGRAMA FA, Partido Nacional: TEXTO DEL PROGRAMA PN,... }
```

```python
# Diccionario para guardar textos en memoria
textos_programas = {}

# Descargar y extraer texto
for url, nombre in zip(urls, nombres_partidos):
    try:
        response = requests.get(url) ## bajar los pdf
        pdf_bytes = BytesIO(response.content)

        texto_completo = "" ## inicializar un texto vacio
        with pdfplumber.open(pdf_bytes) as pdf:
            for page in pdf.pages: # vamos pagina por pagina
                texto_extraido = page.extract_text() # se extrae el texto de esa pagina
                if texto_extraido: #si existe el texto extraido.
                    texto_completo += texto_extraido + "\n" # se agrega a texto_completo; /n salto de linea.

        textos_programas[nombre] = texto_completo # imputamos en el diccionario texto_programas el texto_completo del programa al partido que estamos inspeccionando.
        print(f"Texto extraído correctamente: {nombre}") # Agregar al final de cada loop un print que me indique en que parte del loop va.

    except Exception as e:
        print(f"Error con {nombre}: {e}")

# Arranco con el FA
# genero un texto vacio que se llama texto_completo
# voy a la pagina 1 del FA. Extraigo el texto y lo llamo texto_extraido. Ahora el texto_completo = texto_extraido
# voy a la pagina 2 del FA. Extraigo el texto de la pagina 2 y lo llamo texto_extraido. Ahora el texto_completo = texto_extraido de la pagina 1 + texto extraido de la pagina 2
# voy a la pagina 3 del FA. Extraigo el texto de la pagina 3 y lo llamo texto_extraido. Ahora el texto_completo = texto_completo (hasta ahora, eso es pags 1 y 2) + texto extraido de la pagina 3

# desarrollo.\n
# Vamos
# Bag Of Words ~ Bolsa de Palabras.
```

```python
stop_words = set(stopwords.words('spanish'))

def limpiar_texto(texto):
    tokens = texto.lower().replace('"', '').replace('-', '').replace(',', '').replace('¿', '').replace('.', '').replace('!', '').replace('?', '').split()
    tokens = [palabra for palabra in tokens if palabra.isalpha() and palabra not in stop_words] # solo quedarnos con las palabras que son alphanumericas (letras o numeros, "Hola => No es alphanumerico porque la palabra incluye comillas)
    return tokens

# f(x)=x^2 => Esto funciona como una maquina, a la que yo le tiro un numero y me devuelve el cuadrado de ese numero. Si le doy 2, me devuelve 4.
# f(x)=limpiar_texto => Esto funciona como una maquina, a la que yo le tiro un texto y me devuelve el texto, pero en minusculas, sin signos de puntuacion, y solo con las palabras que son alfanumericas y no son una stop_word

# La funcion actual devuelve tokens_incluidos. Podria generar una lista de tokens excluidos [palabra for palabra in tokens if not palabra.isalpha()]

tokens_programas = {nombre: limpiar_texto(texto) for nombre, texto in textos_programas.items()}

# Ejemplo visual
for partido, tokens in tokens_programas.items():
    print(f"\n{partido} tokens iniciales:", tokens[:10])
```

```python
# Importar librerías necesarias
import requests
import json
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Descargar stopwords en inglés
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

# Descargar catálogo de canciones de The Beatles
url = 'https://raw.githubusercontent.com/skut21x-ga/beatles-api-repo/refs/heads/master/db/lyrics.json'
response = requests.get(url)
data = json.loads(response.text)

# Crear DataFrame y filtrar canciones solo de "The Beatles"
df = pd.DataFrame(data)
df_beatles = df[df['Name'] == 'The Beatles'].reset_index(drop=True)

# Pre-procesar textos

def preprocess(text):
    # Minúsculas
    text = text.lower()
    # Remover puntuación
    text = text.replace('"', '').replace('-', '').replace(',', '').replace('¿', '').replace('.', '').replace('!', '').replace('?', '')
    # Remover stopwords
    text = ' '.join([word for word in text.split() if word not in stop_words])
    return text

# Aplicar preprocesamiento
df_beatles['clean_lyrics'] = df_beatles['Lyrics'].apply(preprocess)

# Vectorización con TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df_beatles['clean_lyrics'])

# Calcular similitud coseno entre canciones
similarity_matrix = cosine_similarity(X)

```

```python
# Encontrar las dos canciones más parecidas
import numpy as np

# Ignorar diagonal (similitud de canciones consigo mismas)
np.fill_diagonal(similarity_matrix, 0)

# Encontrar índices de las canciones más similares
most_similar = np.unravel_index(np.argmax(similarity_matrix), similarity_matrix.shape)

# Mostrar resultados
song1 = df_beatles.iloc[most_similar[0]]['Song']
song2 = df_beatles.iloc[most_similar[1]]['Song']
similarity_score = similarity_matrix[most_similar]

print(f"Las dos canciones más parecidas son '{song1}' y '{song2}' con una similitud de {similarity_score:.2f}")
```

```python
# Modificar la correspondencia de STEM.
from nltk.stem.snowball import SnowballStemmer

stemmer = SnowballStemmer('spanish')

###
# Diccionario con excepciones para casos específicos
manual_stems = {
    'generar': 'gener',   # por ejemplo
    'género': 'genero',
}

# Función que combina SnowballStemmer con override manual
def custom_stem(palabra):
    palabra_lower = palabra.lower()
    return manual_stems.get(palabra_lower, stemmer.stem(palabra)) # esto lo que hace es, para cada palabra, va a buscar primero si la defini manualmente en manual_stems. si hay una definicion manual, va por ella. Sino, va por el stemmer
###

# Integrar el override en tu comprensión de diccionario actual
tokens_stem = {
    partido: [custom_stem(palabra) for palabra in tokens]
    for partido, tokens in tokens_programas.items()
}
```

```python
# Tokens excluidos.
def limpiar_texto(texto):
    tokens = texto.lower().replace('"', '').replace('-', '').replace(',', '').replace('¿', '').replace('.', '').replace('!', '').replace('?', '').split()
    incluidos = [palabra for palabra in tokens if palabra.isalpha() and palabra not in stop_words] ###
    excluidos = [palabra for palabra in tokens if not (palabra.isalpha() and palabra not in stop_words)] ###
    return incluidos, excluidos

tokens_programas = {
    nombre: {
        "incluidos": incluidos,
        "excluidos": excluidos
    }
    for nombre, texto in textos_programas.items()
    for incluidos, excluidos in [limpiar_texto(texto)]
}
```
