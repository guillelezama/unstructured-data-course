```python
import requests
import json
from collections import Counter
import re

# URL
url = 'https://raw.githubusercontent.com/guillelezama/unstructured/refs/heads/main/cfkargentina.json'

# Abrir la JSON data y convertirlo a una lista solo con los tuits (recordar que el último elemento no es un tuit)
response = requests.get(url)
data = json.loads(response.text)[:-1]  # El último elemento no es un tuit, lo descartamos

# Concatenar el texto
all_text = " ".join([item['text'].encode('utf-8').decode('unicode_escape') for item in data])

# Extraer las palabras en una lista, ignorando puntuación
words = re.findall(r'\b\w+\b', all_text.lower())


```

```python
import nltk
from nltk.corpus import stopwords # es una biblioteca que va a bajar la lista de todas las palabras del estilo (articulos, y, me)
# Descargar stopwords en español (solo necesario la primera vez)
nltk.download('stopwords')
# Obtener la lista de stopwords en español
stop_words_es = set(stopwords.words('spanish'))
palabras_filtradas = [X for X in words if X not in stop_words_es and len(X)>3 and X!='http' and X!='https'] #remuevo stop words, palabras de hasta 3 caracteres, http y https que refieren a links


# Generar el contador para cada palabra (usen la lista words)
word_counts = Counter(palabras_filtradas)

# Imprimir LAS PALABRAS 50 MÁS COMUNES una vez filtrado con el stopwords (filtrar y dps usar most_common)
print(word_counts.most_common(50))

```

```python
from wordcloud import WordCloud # para generar la nube
import matplotlib.pyplot as plt # para generar la grafica

# Hacer el wordcloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_counts)
plt.figure(figsize=(12, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
