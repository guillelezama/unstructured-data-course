### 2.3 Ejemplo: Extraer datos del clima usando wttr.in

```python
import requests
import pandas as pd
```

```python
import json
```

```python
ciudades = ["Santiago", "Lima", "Quito", "Trenton", "Montevideo"]

# Lista donde iremos acumulando los diccionarios de resultados
datos_clima = []

# Loop sobre las ciudades
for ciudad in ciudades:
    url = f"https://wttr.in/{ciudad}?format=j1"
    resp = requests.get(url) # Solicitamos datos a un servidor web
    resp.raise_for_status() # si da error, la ejecución se detiene
    js = resp.json() # Convertimos la respuesta en un diccionario de python
    current = js['current_condition'][0] #extraemos el primer elemento de 'current_condition'
    # añadimos a la lista, un diccionario con algunos datos de la ciudad
    datos_clima.append({
        "ciudad": ciudad,
        "temperatura_C": float(current['temp_C']),
        "humedad_%": float(current['humidity']),
        "descripcion": current['weatherDesc'][0]['value'],
    })

# pasamos la lista a un dataframe (tabla)
df_clima = pd.DataFrame(datos_clima)
print(df_clima)
```

## Tarea 2 (4 y 5 dificil!)
1. Abrir el JSON con los tuits de Cristina Fernández de Kirchner. Link: "https://raw.githubusercontent.com/guillelezama/unstructured/refs/heads/main/cfkargentina.json"
2. ¿Qué contenido tiene? Listas? Diccionarios? Diccionarios con listas? etc.
3. ¿Cuántos tuits hay en el archivo?
4. ¿Qué tipos (type) de tuits hay? ¿Cuántos hay de cada uno?

```python
# Tarea

# Importar bibliotecas necesarias
import requests
import json
```

```python
# 1. Abrir el JSON desde la URL dada
url = "https://raw.githubusercontent.com/guillelezama/unstructured/refs/heads/main/cfkargentina.json"
respuesta = requests.get(url)
tuits = respuesta.json()

```

```python
# 2. Explorar la estructura del contenido del JSON
# Recomiendo abrir la url y mirarlo. Podemos explorar el primer elemento
print(type(tuits))  # tipo principal
print(type(tuits[0]))  # tipo del primer elemento
print(tuits[0].keys())  # claves del primer tuit
```

```python
# en este caso, el ultimo elemento no es un tuit. La forma de "ver" el ultimo elemento de una lista es indexarlo con el -1.
print(type(tuits[-1]))  # tipo del primer elemento
print(tuits[-1].keys())  # claves del primer tuit
print((tuits[-1]))
```

```python
# Vemos que el último elemento son características del político
```

```python
# 3. ¿Cuántos tuits hay en total?
num_tuits = len(tuits) - 1 # ya que el ultimo elemento de la list no es un tuit
print(f"Número total de tuits: {num_tuits}")
```

```python
# De hecho, a partir de ahora, saquemos el último elemento de la lista tuits
tuits = tuits[:-1] # quitamos el último elemento de la lista. [:-1] quiere decir, todos los elementos (:) hasta el ultimo (-1)
```

```python
# 4. ¿Qué tipos (type) de tuits hay y cuántos de cada uno?
# Primer observemos los primeros 5 tuits
print(tuits[:5])
```

```python
# type refiere a si es retweet, original, respuesta, etc.

tipos_tuits = {}

for tuit in tuits: # es un loop sobre todos los tuits, es decir, va a repetir esto 13079 veces
    tipo = tuit["type"]  # le digo a la computadora que guarde el valor tipo
    if tipo in tipos_tuits: # si ese tipo ya aparecio antes
        tipos_tuits[tipo] += 1 # el valor del diccionario para ese tipo va a incrementarse en uno
    else:
        tipos_tuits[tipo] = 1 # si ese tipo nunca apareció hasta ahora, lo agrega y pone el valor 1

print("Cantidad de tuits por tipo:")
for tipo, cantidad in tipos_tuits.items():
    print(f"{tipo}: {cantidad}")

```

```python
# 5. Computar el promedio de retweets solo de tuits originales
retweets_totales = 0
contador_originales = 0

for tuit in tuits:
    if tuit['type'] == '"original"': # acá cuidado, porque el valor del tipo es "original" (incluyendo las doble comillas)
        retweets_totales += tuit['metrics']['retweet_count']
        contador_originales += 1

promedio_retweets = retweets_totales / contador_originales if contador_originales > 0 else 0

print(f"Promedio de retweets de tuits originales: {promedio_retweets:.2f}")
```

Tarea: Organizar estos datos (titulo y fecha de publicación en una tabla con formato pandas)

```python
# Tarea

# Importar librerías necesarias
import requests
import xml.etree.ElementTree as ET
import pandas as pd
```

```python
# Descargar el contenido del RSS de BBC
url = "http://feeds.bbci.co.uk/news/rss.xml"
respuesta = requests.get(url)
```

```python
# Parsear contenido XML
raiz = ET.fromstring(respuesta.content)

# Encontrar elementos con noticias (item)
items = raiz.findall(".//item")

# Crear listas para guardar títulos y fechas
titulos = []
fechas = []

# Extraer título y fecha de cada noticia
for item in items:
    titulo = item.find("title").text
    fecha = item.find("pubDate").text

    titulos.append(titulo)
    fechas.append(fecha)

# Crear DataFrame con pandas
df_bbc = pd.DataFrame({"Título": titulos, "Fecha de publicación": fechas})

# Mostrar la tabla
print(df_bbc.head())
```

## Tarea: Parsear el RSS de Noticias de la ONU
link: "https://news.un.org/feed/subscribe/en/news/all/rss.xml"

Realizar una tabla con las siguientes columnas:
- Título
- Fecha

```python
# Tarea

# Importar librerías necesarias
import requests
import xml.etree.ElementTree as ET
import pandas as pd
```

```python
# Descargar contenido del RSS
url = "https://news.un.org/feed/subscribe/en/news/all/rss.xml"
respuesta = requests.get(url)

# Parsear contenido XML
raiz = ET.fromstring(respuesta.content)

# Encontrar elementos con noticias (item)
items = raiz.findall(".//item")

# Crear listas vacías para guardar títulos y fechas
titulos = []
fechas = []

# Extraer título y fecha de cada noticia
for item in items:
    titulo = item.find("title").text
    fecha = item.find("pubDate").text

    titulos.append(titulo)
    fechas.append(fecha)

# Crear DataFrame con pandas
df_noticias = pd.DataFrame({"Título": titulos, "Fecha": fechas})

# Mostrar tabla
print(df_noticias.head())
```
