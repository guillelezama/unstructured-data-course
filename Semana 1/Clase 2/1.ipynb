### 2.3 Ejemplo: Extraer datos del clima usando wttr.in

```python
import requests
import pandas as pd
```

```python
ciudades = ["Buenos Aires", "Santiago", "Lima", "Quito", "Trenton", "Montevideo"]

# Lista donde iremos acumulando los diccionarios de resultados
datos_clima = []

# Loop sobre las ciudades
for ciudad in ciudades:
    url = f"https://wttr.in/{ciudad}?format=j1"
    resp = requests.get(url) # Solicitamos datos a un servidor web
    resp.raise_for_status() # si da error, la ejecución se detiene
    js = resp.json() # Convertimos la respuesta en un diccionario de python
    current = js['current_condition'][0] #extraemos el primer elemento de 'current_condition'
    # añadimos a la lista, un diccionario con algunos datos de la ciudad
    datos_clima.append({
        "ciudad": ciudad,
        "temperatura_C": float(current['temp_C']),
        "humedad_%": float(current['humidity']),
        "descripcion": current['weatherDesc'][0]['value'],
    })

# pasamos la lista a un dataframe (tabla)
df_clima = pd.DataFrame(datos_clima)
print(df_clima)
```

## 3. Trabajando con XML
### 3.1 Introducción a `xml.etree.ElementTree`

```python
import xml.etree.ElementTree as ET
```

```python
# Ejemplo de string XML
xml_str = '''
<ciudades>
  <ciudad nombre="Nueva York">
    <habitantes>8449000</habitantes>
  </ciudad>
  <ciudad nombre="Seattle">
    <habitantes>744900</habitantes>
  </ciudad>
</ciudades>
'''
```

### 3.2 Leer archivos XML

```python
import requests
```

```python
# 1. Definimos la URL del raw XML
url = "https://raw.githubusercontent.com/guillelezama/unstructured/refs/heads/main/ciudades.xml"

# 2. Hacemos la petición y comprobamos que fue exitosa
resp = requests.get(url)
resp.raise_for_status()
```

```python
resp.content
```

```python
# 3. Parseamos el contenido descargado
root = ET.fromstring(resp.content)
```

```python
print(root)
```

```python
print(root.findall('ciudad'))
```

```python
# 4. Recorremos los nodos <ciudad>
for ciudad in root.findall('ciudad'):
    print(ciudad.get('nombre'), ciudad.findtext('habitantes'), ciudad.get('hola'), ciudad.findtext('hola'))
```

```python
# 5. Recorremos los nodos <ciudad>
for hola in root.findall('hola'):
    print(hola.get('nombre'), hola.get('habitantes'), hola.get('hola'))
```

```python
#6. ¿Cuáles son las etiquetas de primer nivel?
print("Hijos directos de la raíz:")
for child in root:
    print(child.tag)
```

```python
#7. ¿Cuáles son todas las etiquetas?
tags = {elem.tag for elem in root.iter()}
print("\nTodas las etiquetas encontradas:", tags)
```

### 3.3 Ejemplo: BBC

```python
import requests
from lxml import etree

# 1. Descargar feed RSS
rss = requests.get("http://feeds.bbci.co.uk/news/rss.xml").content
```

```python
rss
```

```python
root = etree.fromstring(rss)
```

```python
#¿Cuáles son todas las etiquetas únicas?
tags = {elem.tag for elem in root.iter()}
print("\nTodas las etiquetas encontradas:", tags)
```

```python
items = root.xpath('//item') # en cualquier nivel del arbol, buscar todos los que sean items. Xpath es un lenguaje de consulta.
for item in items[:5]:
    title = item.findtext('title')
    pubDate = item.findtext('pubDate')
    print(f"- {title} ({pubDate})")
```
